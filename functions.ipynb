Q1:- **what is the diffrence between a function and a method in python ?*
Ans :-  In Python, the key difference between a function and a method is how they are associated with objects:

1. Function

A function is a block of code that is defined using def or lambda, and it can exist independently of any object.

It is not bound to a class or object unless explicitly passed one.

Example:

def greet(name):
    return f"Hello, {name}!"


2. Method

A method is a function that is associated with an object—usually defined inside a class.

It is called on an object (i.e., an instance of a class) and often takes self as its first parameter to access instance data.

Example:

class Person:
    def greet(self, name):
        return f"Hello, {name}!"

p = Person()
p.greet("Alice")  # `greet` is a method here

Q6:-  explain the concept of generators in python and how do they are define ?

Ans:-  In Python, a generator is a special kind of iterator that allows you to generate values on the fly, instead of storing them all in memory at once.
Key Concepts of Generators

Generators are used to produce a sequence of values, one at a time, using the yield keyword.

They are memory-efficient, especially useful for large datasets or infinite sequences.

Unlike regular functions, a generator remembers its state between calls.
How to Define a Generator

You can define a generator using a function with the yield keyword:

def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

This function returns a generator object:

gen = count_up_to(3)

print(next(gen))  # 1
print(next(gen))  # 2
print(next(gen))  # 3
# next(gen) would raise StopIteration

Differences: yield vs return
Generator Expression (One-liner)

Like list comprehensions, but with () instead of [].

gen = (x * x for x in range(5))
print(next(gen))  # 0

Why Use Generators?

Efficient: No need to store large lists in memory.

Lazy evaluation: Generates items only when requested.

Readable and powerful: Great for pipelines and streaming data.

Q7:-  what are the advantage of using generators over regular functions ?

Ans:-  regular functions—especially when working with large data or infinite sequences. Here are the key benefits:

1. Memory Efficiency

Generators don't store all values in memory; they yield one item at a time.

Regular functions (e.g., ones returning lists) create the entire list in memory at once.


Example:

# Using list (memory heavy)
def get_numbers_list(n):
    return [i for i in range(n)]

# Using generator (memory efficient)
def get_numbers_gen(n):
    for i in range(n):
        yield i

2. Lazy Evaluation

Values are produced only when needed, which makes them ideal for pipelines and streaming data.


gen = (i * i for i in range(1000000))
print(next(gen))  # Only calculates the first value
3. Infinite Sequences

You can easily model infinite series with generators, which is not possible with lists.


def count_forever():
    i = 0
    while True:
        yield i
        i += 1

4. Improved Performance in Loops

Since generators yield items one at a time, looping through them is faster and less resource-intensive.

5. Cleaner Code for Iteration

When using yield, the code for custom iterators becomes simpler and more readable than writing classes with __iter__() and __next__() methods.

Q10:- what is diffrence between map(); reduce(); and filter(); function in python?

Ans:-  In Python, map(), filter(), and reduce() are functional programming tools that work on iterables. Here's a breakdown of the differences between them:


---

1. map()

Purpose: Applies a function to each item in an iterable.

Returns: A new iterable with the transformed values.


# Example: Square each number
numbers = [1, 2, 3]
result = map(lambda x: x**2, numbers)
print(list(result))  # [1, 4, 9]


---

2. filter()

Purpose: Filters elements based on a condition (function that returns True or False).

Returns: A new iterable with only the items that pass the test.


# Example: Keep only even numbers
numbers = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))  # [2, 4]


---

3. reduce() (from functools module)

Purpose: Applies a function cumulatively to the items in an iterable, reducing it to a single value.

Returns: A single aggregated result.


from functools import reduce

# Example: Multiply all numbers
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x * y, numbers)
print(result)  # 24

Q.5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

Ans.
Here's a Python iterator class that generates the Fibonacci sequence up to a specified number of terms:

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage
fib = FibonacciIterator(10)
for num in fib:
    print(num)

3
5
8
13
21
34

This class:

Initializes with the number of terms desired.

Implements __iter__() and __next__() methods to make it an iterator.

Raises StopIteration once the desired number of terms is reached.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

Q.7.Implement a generator function that reads a file line by line and yields each line as a string.

Ans.
Here's a simple Python generator function that reads a file line by line and yields each line as a string:

def read_file_line_by_line(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Removes trailing newline character

Usage example:

for line in read_file_line_by_line('example.txt'):
    print(line)

This approach is memory-efficient, especially useful for reading large files, as it doesn't load the entire file into memory.
